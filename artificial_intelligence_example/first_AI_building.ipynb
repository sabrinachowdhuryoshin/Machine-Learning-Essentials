#### *Build an AI to detect whether a tumor is malignant (cancerous) or benign (noncancerous)*
---

##### 1. Load the data

```python
import pandas as pd

cancer_data = pd.read_csv('cancer.csv')
cancer_data
```

##### 2. Segregate the input and output data sets

```python
''' segregate the input and output data sets '''

## input data: everything in the dataset except for the diagnosis column
## ouput data: diagnosis column, prediction will be done based on this column
input_data = cancer_data.drop(columns = ['diagnosis(1=m, 0=b)'])
output_data = cancer_data['diagnosis(1=m, 0=b)']

print(input_data)
print(output_data)
```

##### 3. Split the Data into 'Training' and 'Testing' sets

```python
## general rule of thumb: 
## allocating 70-80% of data for training
## alloctaiong 20-30% of data for testing
## so, here a test set will be used for prediction
## then the prediction will be compared with the original data of the test set

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

## train_test_split gives a tuple
input_train, input_test, output_train, output_test = train_test_split(input_data, output_data, test_size = 0.2) ## 0.2 means 20% data is being used for testing
```

##### 4. Create a model using a tensorflow and keras

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
```

##### 5. Create a neural network

```python
model.add(tf.keras.layers.Dense(256, input_shape = input_train.shape, activation = 'sigmoid')) ## 256 is the power here
model.add(tf.keras.layers.Dense(256, activation = 'sigmoid'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid')) # 1 because we have only one output either 1 or 0
```

##### 6. Compile the model and fit the data

```python
## optimizer helps to fine-tune the parameters to fit the data
## loss = 'binary_crossentropy', because we are using binary classification (0 or 1)
## metrics = ['accuracy'], because we want to classify whether the tumor is cancerour or not
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy']) 

## fit the data
## epochs = 1000 means 1000 times it will iterate to fit the data
model.fit(input_train, output_train, epochs = 1000) 
```
